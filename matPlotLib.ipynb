# MatplotLib 

```python
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)

a = np.arange(40,50)
b = np.arange(50,60)

print(x,y,a,b)
```

```python
##plt Scatter

plt.scatter(x,y,c = 'b')  # c for colour
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('2D Scatter Graph')
plt.savefig('Matplot figures/fig1.jpg')
```

```python
#plt Plot
y = x*x
plt.plot(x,y,'r*--',linestyle = 'dashed',linewidth = 2,markersize = 12)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('2D Plot Graph')
plt.savefig('Matplot figures/fig2.jpg')
```

```python
#Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'b*--')
plt.subplot(2,2,3)
plt.plot(x,y,'go')
plt.savefig('Matplot figures/fig3.jpg')
```

```python
#Complex functions

x = np.arange(1,11)
y = 3 * x + 5
plt.title("MatplotLib Demo")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot(x,y)
plt.savefig('Matplot figures/fig4.jpg')
```

```python
#Sin Waves 

x = np.arange(0,4 * np.pi, 0.1)
print(x)
y = np.sin(x)
print(y)
plt.title("SinWave")

plt.plot(x,y,'b*')
plt.savefig('Matplot figures/fig5.jpg')
```

```python
#SubPlot of Sine and Cosine

x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'b--')
plt.title("Sine Wave")

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title("cosine Wave")

plt.savefig('Matplot figures/fig6.jpg')
```

```python
#Bar plot 

x1 = [2,8,10]
y1 = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x1,y1,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graphs')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.savefig('Matplot figures/fig7.jpg')
```

```python
#Histograms

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.savefig('Matplot figures/fig8.jpg')
```

```python
#Box Plot

data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert = True,patch_artist = True)  # Vert is for Vertical position and patch_artist is for colour
plt.savefig('Matplot figures/fig9.jpg')
```

```python
#Pie chart

labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','green','lightskyblue']
explode = (0.1,0,0,0) #explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title("3D Pie Chart")
plt.axis('equal')
plt.savefig('Matplot figures/fig9.jpg')
```
