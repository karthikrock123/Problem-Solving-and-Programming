# SeaBorn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

# Corelation with HeatMap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# JoinPolt

```python
# Univariate Analysis(2 Values)

sns.jointplot(x = 'tip',y = 'size',data = df,kind = 'hex')
sns.jointplot(x = 'tip',y = 'total_bill',data = df,kind = 'reg')
```

```python
sns.pairplot(df)
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue = 'sex')
```

```python
sns.pairplot(df,hue = 'smoker')
```

# DistPlot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde = False,bins = 10)
```

# Categorical Plots

```python
##Count Plot

sns.countplot('smoker',data = df)
```

```python
sns.countplot(y = 'sex',data = df)
```

```python
## Bar Plot

sns.barplot(x = 'sex',y = 'total_bill',data = df)
```

```python
## Box Plot

sns.boxplot('smoker','total_bill',data = df)
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df, palette = 'rainbow')
```

```python
sns.boxplot(data = df,orient = 'v')
```

```python
sns.boxplot(x = 'total_bill',y = 'day',hue = 'smoker' ,data = df)
```

# Violin Plot

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df, palette = 'rainbow')
```
